```python
import pandas as pd
import numpy as np
import re
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df1=pd.read_csv('news.csv')
df2=pd.read_csv('stocks.csv')
df1.head()
```

```python
df1.shape
```

```python
df2.head()
```

```python
df2.shape
```

```python
#Merging The datasets on basis of date
merge=df1.merge(df2,how='inner',on='Date',left_index=True)
```

```python
merge.head()
```

```python
#Combining top news headlines
headlines=[]

for row in range(0, len(merge.index)):
    headlines.append(' '.join(str(x) for x in merge.iloc[row,2:27]))
```

```python
#printing samples
headlines[0]
```

```python
#Cleaning Data
clean_headlines=[]

for i in range(0,len(headlines)):
    clean_headlines.append(re.sub("b[(')]",' ',headlines[i])) #removes b
    clean_headlines[i]=re.sub('b[(")]',' ',clean_headlines[i])
    clean_headlines[i]=re.sub("\'",' ',clean_headlines[i])
```

```python
#COMBINED CLEAN HEADLINES
clean_headlines[0]
```

```python
#adding clean headlines to the dataset
merge['Combined_News']=clean_headlines
merge['Combined_News'][0]
```

```python
merge.head()
```

```python
#Function to get the subjectivity and polarity

def getSubjective(text):
    return TextBlob(text).sentiment.subjectivity

def getpolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
merge['Subjectivity']=merge['Combined_News'].apply(getSubjective)
merge['Polarity']=merge['Combined_News'].apply(getpolarity)
```

```python
merge.head()
```

```python
#creating a Function to get Sentiment Scores
def getSIA(text):
    sia=SentimentIntensityAnalyzer()
    sentiment=sia.polarity_scores(text)
    return sentiment
```

```python
#Grtting Sentiment scores for each day
compound=[]
neg=[]
pos=[]
neu=[]
SIA=0
for i in range(0, len(merge['Combined_News'])):
    SIA=getSIA(merge['Combined_News'][i])
    compound.append(SIA['compound'])
    neg.append(SIA['neg'])
    neu.append(SIA['neu'])
    pos.append(SIA['pos'])
```

```python
merge['Compound']=compound
merge['Negative']=neg
merge['Neutral']=neu
merge['Positive']=pos
```

```python
merge.head()
```

```python
#Create a lit of columns
Keep_columns=['Open','High','Low','Volume','Subjectivity','Polarity','Compound','Negative','Positive','Neutral','Label']
```

```python
df=merge[Keep_columns]
df
```

```python
#Creating a feature dataset
X=df
X=np.array(X.drop('Label', axis=1))

y=df['Label']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
model=LinearDiscriminantAnalysis().fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
predictions
```

```python
y_test
```

```python
#Model Metric
print(classification_report(y_test,predictions))
```
